## Métodos auxiliares
Alguno de los métodos de esta parte se han tomado de los apuntes del campus y han sido modificados. En el caso de los modificados se han indicado los cambios realizados con comentarios (estos cambios están en el apartado Cambios realizados de cada metodo).

```python
import sqlite3
from sqlite3 import IntegrityError

base_de_datos = "Libreria.sqlite3"

tabla_compradores = "Compradores"
tabla_libros = "Libros"
tabla_compras = "Compras"

def makedicts(cursor, query, params=()):
    """Ejecuta una consulta y devuelve los datos almacenados en un diccionario.
    
    Parámetros de entrada:
    cursor --- puntero en la base de datos
    query --- consulta a realizar
    
    Retorno:
    rowdict --- diccionario con los datos de la consulta
    """
    
    cursor.execute(query, params)
    colnames = [desc[0] for desc in cursor.description]
    rowdicts = [dict(zip(colnames, row)) for row in cursor.fetchall()]
    return rowdicts

def showformat(recs, sept=('-' * 40)):
    """Muestra por pantalla los datos pasados como arguentos de entrada.
    
    Parámetros de entrada:
    recs --- diccionario con los datos a mostrar
    """
    
    print(len(recs), 'registros')
    print(sept)
    for rec in recs:
        maxkey = max(len(key) for key in rec)
        for key in rec: # or: \t align
            print('%-*s => %s' % (maxkey, key, rec[key]))
        print(sept)

def mostrar_tabla(cursor, tabla, format=True):
    """Muestra el contenido de una tabla.
    
    Parámetros de entrada:
    cursor --- puntero en la base de datos
    tabla --- tabla a ser mostrada
    """
    
    if not format:
        cursor.execute('select * from ' + tabla)
        while True:
            rec = cursor.fetchone()
            if not rec: break
            print(rec)
    else:
        recs = makedicts(cursor, 'select * from ' + tabla)
        showformat(recs)
        
def vaciar_tabla(cursor, tabla, conexion):
    """Vacía la tabla indicada.
    
    Parámetros de entrada:
    cursor --- puntero en la base de datos
    tabla --- tabla a vaciar
    conexion --- conexion a la bbdd para confirmar cambios
    
    Cambios realizados (apartado Script 5 del archivo ScriptPythonSQL(SQLite).pdf):
    confirmación de los cambios en la bbdd trás vaciar la tabla
    mensaje informativo para el usuario de los cambios llevados a cabo
    """
    
    cursor.execute('DELETE FROM ' + tabla)
    conexion.commit()
    print('\nSe ha vaciado la tabla', tabla)
    
def eliminar_tabla(cursor, base_de_datos, tabla, conexion):
    """Elimina la tabla indicada.
    
    Parámetros de entrada:
    cursor --- puntero en la base de datos
    base_de_datos --- nombre de la base de datos usada
    tabla --- tabla a eliminar
    conexion --- conexion a la bbdd para confirmar cambios
    
    Cambios realizados (apartado Script 4 del archivo ScriptPythonSQL(SQLite).pdf):
    confirmación de los cambios en la bbdd trás vaciar la tabla
    mensaje informativo para el usuario de los cambios llevados a cabo
    """
    
    cursor.execute('DROP TABLE IF EXISTS ' + tabla)
    conexion.commit()
    print('Se ha eliminado la tabla', tabla, 'de la base de datos', base_de_datos)

def eliminar_tablas(cursor, conexion):
    """Vacía y elimina todas las tablas de la base de datos.
    
    Parámetros de entrada:
    cursor --- puntero en la base de datos
    conexion --- conexion a la bbdd para confirmar cambios
    """
    
    vaciar_tabla(cursor, tabla_compradores, conexion)
    eliminar_tabla(cursor, base_de_datos, tabla_compradores, conexion)
    
    vaciar_tabla(cursor, tabla_libros, conexion)
    eliminar_tabla(cursor, base_de_datos, tabla_libros, conexion)
    
    vaciar_tabla(cursor, tabla_compras, conexion)
    eliminar_tabla(cursor, base_de_datos, tabla_compras, conexion)
```

## Ejercicio 1
Crear un programa en Python que genere utilizando SQLite una base de datos denominada “Libreria”  y 3 tablas: compradores, libros y compras [1,5 puntos].

```python
def generardb(base_de_datos):
    """Crea la conexión a la base de datos.
    
    Parámetros de entrada:
    base_de_datos --- nombre de la base de datos
    
    Cambios realizados (apartado Script 2 del archivo ScriptPythonSQL(SQLite).pdf):
    mensaje informativo para el usuario de los cambios llevados a cabo
    """
    
    conexion = sqlite3.connect(base_de_datos)
    cursor = conexion.cursor()
    print('Se ha generado la base de datos', base_de_datos)
    return cursor, conexion

def crear_tabla_compradores(cursor, conexion):
    """Crea la tabla compradores de la base de datos.
    
    Parámetros de entrada:
    cursor --- puntero en la base de datos
    conexion --- conexion a la bbdd para confirmar cambios
    """

    cursor.execute("""
        CREATE TABLE Compradores(
            registro INTEGER(4) PRIMARY KEY NOT NULL UNIQUE,
            nombre VARCHAR(35) NOT NULL DEFAULT '' UNIQUE,
            fecha_nacimiento DATE NOT NULL DEFAULT '0000-00-00',
            telefono VARCHAR(10) DEFAULT NULL,
            domicilio VARCHAR(35) DEFAULT NULL,
            poblacion VARCHAR(25) DEFAULT NULL,
            anotaciones TEXT
        )""")
    conexion.commit()
    print("Se ha creado la tabla Compradores")

    
def crear_tabla_libros(cursor, conexion):
    """Crea la tabla libros de la base de datos.
    
    Parámetros de entrada:
    cursor --- puntero en la base de datos
    conexion --- conexion a la bbdd para confirmar cambios
    """
    
    cursor.execute("""
        CREATE TABLE Libros(
            registro INTEGER(4) NOT NULL PRIMARY KEY UNIQUE,
            titulo VARCHAR(35) NOT NULL DEFAULT '' UNIQUE,
            escritor VARCHAR(35) NOT NULL DEFAULT '',
            editorial VARCHAR(20) NOT NULL DEFAULT '',
            soporte VARCHAR(20) NOT NULL DEFAULT 'Libro',
            fecha_entrada DATE NOT NULL DEFAULT NULL,
            pais VARCHAR(20) NOT NULL DEFAULT NULL,
            importe DECIMAL(8,2) NOT NULL DEFAULT 0.0,
            anotaciones BLOB
        )""")
    conexion.commit()
    print("Se ha creado la tabla Libros")
    
def crear_tabla_compras(cursor, conexion):
    """Crea la tabla compras de la base de datos.
    
    Parámetros de entrada:
    cursor --- puntero en la base de datos
    conexion --- conexion a la bbdd para confirmar cambios
    """
    
    cursor.execute("""
        CREATE TABLE Compras(
            registro INT(4) NOT NULL PRIMARY KEY UNIQUE,
            id_comprador INT(4) NOT NULL DEFAULT '',
            id_libro INT(4) NOT NULL DEFAULT '',
            FOREIGN KEY (id_comprador) REFERENCES Compradores(registro),
            FOREIGN KEY (id_libro) REFERENCES Libros(registro)
        )""")
    conexion.commit()
    print("Se ha creado la tabla Compras")

    
def generar_tablas(cursor, conexion):
    """Crea las tablas de la base de datos.
    
    Parámetros de entrada:
    cursor --- puntero en la base de datos
    conexion --- conexion a la bbdd para confirmar cambios
    """
    # Querys para saber si una tabla existe en la bbdd o no
    query_existe_compradores = """SELECT name FROM sqlite_master WHERE type='table' AND name='Compradores';"""
    query_existe_libros = """SELECT name FROM sqlite_master WHERE type='table' AND name='Libros';"""
    query_existe_compras = """SELECT name FROM sqlite_master WHERE type='table' AND name='Compras';"""

    """Miramos si la tabla compradores existe en la bbdd (si la longitud del diccionario devuelto
    por makedicts es 0 no tenemos dicha tabla). Si no existe la creamos y en ambos casos informamos
    al usuario de los resultados. Repetimos el mismo proceso para el resto de tablas."""
    if (len(makedicts(cursor, query_existe_compradores)) == 0):
        print("\nLa tabla Compradores no existe en la base de datos.")
        crear_tabla_compradores(cursor, conexion)
    else:
        print("\nLa tabla Compradores existe en la base de datos.")

    if (len(makedicts(cursor, query_existe_libros)) == 0):
        print("\nLa tabla Libros no existe en la base de datos.")
        crear_tabla_libros(cursor, conexion)
    else:
        print("\nLa tabla Libros existe en la base de datos.")

    if (len(makedicts(cursor, query_existe_compras)) == 0):
        print("\nLa tabla Compras no existe en la base de datos.")
        crear_tabla_compras(cursor, conexion)

# METODO SOLICITADO POR EL ENUNCIADO
def inicializardb(base_de_datos):
    """Estalece la conexion con la base de datos y crea las tablas.
    
    Parámetros de entrada:
    base_de_datos --- nombre de la base de datos usada
    """
    
    cursor, conexion = generardb(base_de_datos)
    generar_tablas(cursor, conexion)
    return cursor, conexion
```

## Ejercicio 2
Crear un programa en Python que rellene las tablas anteriores [1 punto].

```python
def cargar_tabla(cursor, tabla, fichero_tabla, conexion, verbose=True):
    """Carga la información de una tabla de la bbdd desde un archivo.
    
    Parámetros de entrada:
    cursor --- puntero en la base de datos
    tabla --- tabla sobre la que realizar insercciones
    fichero_tabla --- fichero sobre el que extraer la información a cargar
    conexion --- conexion a la bbdd para confirmar cambios
    
    Cambios realizados (apartado Script 2 del archivo ScriptPythonSQL(SQLite).pdf):
    mensaje informativo para el usuario de los cambios llevados a cabo
    mensaje informativo en caso de que alguna tupla no sea valida
    secuencia try-except por si alguna tupla a insertar no es valida
    """
    
    with open(fichero_tabla, 'r', encoding="utf8") as file:# x,x,x\nx,x,x\n
        rows = [line.rstrip().split('|') for line in file] # [[x,x,x], [x,x,x]]
        rows = [str(tuple(rec)) for rec in rows] # ["(x,x,x)", "(x,x,x)"]
        for recstr in rows:
            print(recstr)
            try:
                cursor.execute('INSERT INTO ' + tabla + ' VALUES ' + recstr)
            except IntegrityError:
                print("Problema de integridad con los datos introducidos")
        conexion.commit()
        if verbose: 
            print (len(rows), 'filas cargadas en la tabla', tabla, 'desde el fichero', fichero_tabla)

# METODO SOLICITADO POR EL ENUNCIADO
def llenar_tablas(cursor, conexion):
    """Carga la información de las tablas de la bbdd.
    
    Parámetros de entrada:
    cursor --- puntero en la base de datos
    conexion --- conexion a la bbdd para confirmar cambios
    """
    
    # Querys para comprobar si las tablas estan vacias o contienen informacion
    query_tbcompradores_vacia = """SELECT * FROM Compradores;"""
    query_tblibros_vacia = """SELECT * FROM Libros;"""
    query_tbcompras_vacia = """SELECT * FROM Compras;"""

    # Ficheros con los que cargar las tablas
    fichero_compradores = "compradores.txt"
    fichero_libros = "libros.txt"
    fichero_compras = "compras.txt"
    
    """Miramos si la tabla compradores esta vacia (si la longitud del diccionario devuelto
    por makedicts es 0 no esta vacia). Si no lo esta la vaciamos. Lo este o no informamos
    al usuario de los resultados y procedemos a llenarla. Repetimos el mismo proceso para 
    el resto de tablas."""
    if (len(makedicts(cursor, query_tbcompradores_vacia)) != 0):
        print("\nLa tabla Compradores contiene información.")
        vaciar_tabla(cursor, tabla_compradores, conexion)
    else:
        print("\nLa tabla Compradores está vacía.")
    cargar_tabla(cursor, tabla_compradores, fichero_compradores, conexion)

    if (len(makedicts(cursor, query_tblibros_vacia)) != 0):
        print("\nLa tabla Libros contiene información.")
        vaciar_tabla(cursor, tabla_libros, conexion)
    else:
        print("\nLa tabla Libros está vacía.")
    cargar_tabla(cursor, tabla_libros, fichero_libros, conexion)

    if (len(makedicts(cursor, query_tbcompras_vacia)) != 0):
        print("\nLa tabla Compras contiene información.")
        vaciar_tabla(cursor, tabla_compras, conexion)
    else:
        print("\nLa tabla Compras está vacía.")
    cargar_tabla(cursor, tabla_compras, fichero_compras, conexion)        
```

## Ejercicio 3
Crear un programa en Python que resuelva las siguientes consultas SQL [7,5 puntos].

* Obtener los países y el número de libros vendidos agrupados por país y ordenados de manera descendente respecto al total de ventas.

```python
# METODO SOLICITADO POR EL ENUNCIADO
def query_libros_vendidos_por_pais(cursor):
    """Ejecuta la consulta 1 del enunciado y muestra los resultados por pantalla.
    
    Parámetros de entrada:
    cursor --- puntero en la base de datos
    """
    
    print('\n\n\nCONSULTA 1')
    query = """
        SELECT pais AS [País], COUNT(pais) AS [Numero de libros], ROUND(SUM(importe),3) AS [Importe total]
        FROM Libros lib, Compras com
        WHERE lib.registro = com.id_libro
        GROUP BY pais
        ORDER BY [Importe total] DESC
        """
    
    showformat(makedicts(cursor, query))
```

* Obtener el nombre de los compradores que han comprado al menos un libro y número de libros comprados, ordenados decrecientemente por el número total de libros comprados.

```python
# METODO SOLICITADO POR EL ENUNCIADO
def query_libros_comprados_ord_dec(cursor):
    """Ejecuta la consulta 2 del enunciado y muestra los resultados por pantalla.
    
    Parámetros de entrada:
    cursor --- puntero en la base de datos
    """
    
    print('\n\n\nCONSULTA 2')    
    query = """
        SELECT nombre AS [Nombre], count(*) AS [Numero de libros]
        FROM Compras, Compradores
        WHERE Compras.id_comprador = Compradores.registro
        GROUP BY nombre
        ORDER BY [Numero de libros] DESC
        """
    
    showformat(makedicts(cursor, query))
```

* Obtener la media de lo que han gastado todos los compradores y la suma total de todas las ventas realizadas.

```python
# METODO SOLICITADO POR EL ENUNCIADO
def query_media_gastos_suma_ventas(cursor):
    """Ejecuta la consulta 3 del enunciado y muestra los resultados por pantalla.
    
    Parámetros de entrada:
    cursor --- puntero en la base de datos
    """
    
    print('\n\n\nCONSULTA 3')
    # Aplicamos ROUND por si al hacer la division quedan demasiados decimales
    query = """
        SELECT SUM(importe) AS [Suma total], ROUND(SUM(importe)/(SELECT COUNT(registro) 
                                                           FROM Compradores),2) AS [Media por cliente]
        FROM Libros, Compras
        WHERE Libros.registro == Compras.id_libro"""
    
    showformat(makedicts(cursor, query))
```

```python
# METODO SOLICITADO POR EL ENUNCIADO
def query_info_compradores(cursor):
    """Ejecuta la consulta 4 del enunciado y muestra los resultados por pantalla.
    
    Parámetros de entrada:
    cursor --- puntero en la base de datos
    """
    
    print('\n\n\nCONSULTA 4')
    query = """
        SELECT nombre AS [Nombre], telefono AS [Teléfono], anotaciones AS [Anotaciones]
        FROM Compradores
        WHERE nombre LIKE 'M%'"""
    
    showformat(makedicts(cursor, query))
```

* Actualizar la tabla Libros, cambiando el soporte de los registros 6 y 7 a “DVD”.

```python
# METODO SOLICITADO POR EL ENUNCIADO
def query_actualiza_libros(cursor, conexion):
    """Ejecuta la consulta 5 del enunciado y muestra los resultados por pantalla.
    
    Parámetros de entrada:
    cursor --- puntero en la base de datos
    conexion --- conexion a la bbdd para confirmar cambios
    """

    print('\n\n\nCONSULTA 5')
    query1 = """
        SELECT registro AS [Registro], titulo AS [Título], soporte AS [Soporte]
        FROM Libros
        WHERE registro==6 OR registro==7"""
        
    query2 = """
        UPDATE Libros
        SET soporte='DVD'
        WHERE registro==6 OR registro==7"""
    
    query3 = """
        SELECT registro AS [Registro], titulo AS [Título], soporte AS [Soporte]
        FROM Libros
        WHERE registro==6 OR registro==7"""
    
    # Mostramos por pantalla los datos, antes y despues de ejecutar la query, para ver los cambios realizados
    print('\nAntes de actualizar la tabla Libros')
    showformat(makedicts(cursor, query1))
    cursor.execute(query2)
    conexion.commit()
    print('\nDespués de actualizar la tabla Libros')
    showformat(makedicts(cursor, query3))
```

* Borrar los libros que no se han vendido nunca.

```python
# METODO SOLICITADO POR EL ENUNCIADO
def query_elimina_libros_no_vendidos(cursor, conexion):
    """Ejecuta la consulta 6 del enunciado y muestra los resultados por pantalla.
    
    Parámetros de entrada:
    cursor --- puntero en la base de datos
    conexion --- conexion a la bbdd para confirmar cambios
    """
    
    print('\n\n\nCONSULTA 6')
    
    query1 = """
        DELETE FROM Libros
        WHERE registro NOT IN (
            SELECT id_libro
            FROM Compras)"""
    
    query2 = """
        SELECT registro AS [Registro], titulo AS [Título]
        FROM Libros"""

    # Mostramos por pantalla los datos, antes y despues de ejecutar la query, para ver los cambios realizados
    print('\nAntes de actualizar la tabla Libros')
    showformat(makedicts(cursor, query2))
    cursor.execute(query1)
    conexion.commit()
    print('\nDespués de actualizar la tabla Libros')
    showformat(makedicts(cursor, query2))
```
